```python
from collections import Counter


def load_data(path):
    with open(path) as f:
        data = f.read().splitlines()
    return data


def parse_data(data):
    left_list = []
    right_list = []
    for row in data:
        left_value, right_value = row.split()
        left_list.append(int(left_value))
        right_list.append(int(right_value))
    left_list = sorted(left_list)
    right_list = sorted(right_list)
    return left_list, right_list


def solve_part_1(left_list, right_list):
    total_distance = 0
    for i in range(len(left_list)):
        left_value = left_list[i]
        right_value = right_list[i]
        total_distance += abs(left_value - right_value)
    return total_distance


def solve_part_2(left_list, right_list):
    total_similarity = 0
    counter_rl = Counter(right_list)
    for _, left_value in enumerate(left_list):
        total_similarity += left_value * counter_rl[left_value]
    return total_similarity


data = load_data('input.txt')
left_list, right_list = parse_data(data)
print(solve_part_1(left_list, right_list))
print(solve_part_2(left_list, right_list))
```
