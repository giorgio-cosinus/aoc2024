```python
import re
 

with open('input.txt') as f:
    data = f.read()
 
sections = re.findall(r'mul\((\d+,\d+)\)', data)

total_corrupted_memory = 0
for section in sections:
    x, y = [int(number) for number in section.split(',')]
    total_corrupted_memory += x * y

total_corrupted_memory
```

```python
import re


with open('input.txt') as f:
    data = f.read()

sections_raw = re.findall(r"mul\((\d+,\d+)\)|(do\(\))|(don't\(\))", data)

sections = []
for section in sections_raw:
    for instruction in section:
        if instruction:
            sections.append(instruction)

is_enabled = True
total_corrupted_memory = 0
for section in sections:
    if section == "don't()":
        is_enabled = False
    elif section == 'do()':
        is_enabled = True
    elif is_enabled is True:
        x, y = [int(number) for number in section.split(',')]
        total_corrupted_memory += x * y

total_corrupted_memory
```
