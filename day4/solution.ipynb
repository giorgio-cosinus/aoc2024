```python
with open('input.txt') as f:
    data = f.read().splitlines()
```

```python
def read_text_horizontal(grid, coordinates):
    x, y = coordinates
    text = grid[x][y:y+4]
    return text


def read_text_vertical(grid, coordinates):
    x, y = coordinates
    text = ''.join([i[y] for i in data[x:x+4]])
    return text


def read_text_diagonal_positive(grid, coordinates):
    x, y = coordinates
    text = ''.join([v[y+i] for i, v in enumerate(data[x:x+4])])
    return text


def read_text_diagonal_negative(grid, coordinates):
    x, y = coordinates
    text = ''.join([v[y-i] for i, v in enumerate(data[x:x+4])])
    return text


def is_text_xmas(text):
    if text == 'XMAS' or text == 'SAMX':
        return True
    else:
        return False
```

```python
def solve_part_1(grid):
    total_rows = len(grid)
    total_cols = len(grid[0])
    xmas_occurences = 0
    for i in range(total_rows):
        for j in range(total_cols):
            xmas_occurences += is_text_xmas(read_text_horizontal(grid, (i, j)))
            xmas_occurences += is_text_xmas(read_text_vertical(grid, (i, j)))
            if i <= total_rows - 4 and j <= total_cols - 4:
                xmas_occurences += is_text_xmas(read_text_diagonal_positive(grid, (i, j)))
            if i <= total_rows - 4 and j >= 3:
                xmas_occurences += is_text_xmas(read_text_diagonal_negative(grid, (i, j)))
    return xmas_occurences


solve_part_1(data)
```

```python
def rotate_shape(shape):
    rotated_shape = [''.join(list(reversed(i))) for i in zip(*shape)]
    return rotated_shape


def shape_to_x_mas_text(shape):
    text = shape[0][0] + shape[0][2] + shape[1][1] + shape[2][0] + shape[2][2]
    return text


def is_shape_x_mas(shape):
    shape_90 = rotate_shape(shape)
    shape_180 = rotate_shape(shape_90)
    shape_270 = rotate_shape(shape_180)
    reference_shape = ['M.S', '.A.', 'M.S']

    text = shape_to_x_mas_text(shape)
    text_90 = shape_to_x_mas_text(shape_90)
    text_180 = shape_to_x_mas_text(shape_180)
    text_270 = shape_to_x_mas_text(shape_270)
    reference_text = shape_to_x_mas_text(reference_shape)

    if text == reference_text:
        return True
    elif text_90 == reference_text:
        return True
    elif text_180 == reference_text:
        return True
    elif text_270 == reference_text:
        return True
    else:
        return False
```

```python
def solve_part_2(grid):
    total_rows = len(grid)
    total_cols = len(grid[0])
    x_mas_occurences = 0
    for i in range(total_rows-2):
        for j in range(total_cols-2):
            shape = [grid[i][j:j+3], grid[i+1][j:j+3], grid[i+2][j:j+3]]
            x_mas_occurences += is_shape_x_mas(shape)
    return x_mas_occurences


solve_part_2(data)
```
