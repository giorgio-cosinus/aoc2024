```python
import re


def parse_input(data):
    wires_, operations_ = data
    wires_ = wires_.split('\n')
    operations_ = operations_.split('\n')
    wires = {}
    for wire_ in wires_:
        variable, value = re.findall(r'(\w+): (\d+)', wire_)[0]
        wires[variable] = int(value)
    operations = []
    for operation_ in operations_:
        operation = re.findall(r'(\w+) (\w+) (\w+) -> (\w+)', operation_)[0]
        operations.append(list(operation))
    return wires, operations


def get_wire_source(wire, operations):
    for i, operation in enumerate(operations):
        if wire == operation[3]:
            return i


def evaluate_bitwise(x, op, y):
    match op:
        case 'AND':
            return x & y
        case 'OR':
            return x | y
        case 'XOR':
            return x ^ y


def connect_wires(wires, operations):
    operations = operations.copy()
    while operations:
        x, op, y, z = operations[0]
        if x in wires and y in wires:
            wires[z] = evaluate_bitwise(wires[x], op, wires[y])
        elif x not in wires:
            index = get_wire_source(x, operations)
            src_element = operations.pop(index)
            operations.insert(0, src_element)
            continue
        elif y not in wires:
            index = get_wire_source(y, operations)
            src_element = operations.pop(index)
            operations.insert(0, src_element)
            continue
        operations.pop(0)
    return dict(sorted(wires.items(), reverse=True))


def solve_part_2(path):
    with open(path) as f:
        data = f.read().split('\n\n')
        wires, operations = parse_input(data)
        connected_wires = connect_wires(wires, operations)
        highest_z = list(connected_wires)[0]
    swapped_outputs = set()
    for operation in operations:
        x, op, y, z = operation
        if op == 'XOR' and x[0] not in 'xy' and y[0] not in 'xy' and z[0] != 'z':
            swapped_outputs.add(z)
        elif op != 'XOR' and z[0] == 'z' and z != highest_z:
            swapped_outputs.add(z)
        elif op == 'XOR':
            for x_, op_, y_, z_ in operations:
                if (z in [x_, y_]) and op_ == 'OR':
                    swapped_outputs.add(z)
        elif op == 'AND' and 'x00' not in [x, y]:
            for x_, op_, y_, z_ in operations:
                if (z in [x_, y_]) and op_ != 'OR':
                    swapped_outputs.add(z)
    return ','.join(sorted(swapped_outputs))


def solve_part_1(path):
    with open(path) as f:
        data = f.read().split('\n\n')
        wires, operations = parse_input(data)
        connected_wires = connect_wires(wires, operations)
    binary_number = ''
    for wire, value in connected_wires.items():
        if wire.startswith('z'):
            binary_number += str(value)
    decimal_number = int(binary_number, 2)
    return decimal_number


print(solve_part_1('input.txt'))
print(solve_part_2('input.txt'))
```
