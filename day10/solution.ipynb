```python
def int_(char):
    if char == '.':
        return -1
    return int(char)


def dfs(board, board_shape, coordinates, visited_coordinates):
    x, y = coordinates
    if board[x][y] == '.':
        value = -1
    else:
        value = int_(board[x][y])
    neighbors = []
    if coordinates not in visited_coordinates:
        visited_coordinates.add(coordinates)
        if x >= 1 and value == int_(board[x-1][y]) - 1:
            neighbors.append((x-1, y))
        if y <= board_shape[1] - 2 and value == int_(board[x][y+1]) - 1:
            neighbors.append((x, y+1))
        if x <= board_shape[0] - 2 and value == int_(board[x+1][y]) - 1:
            neighbors.append((x+1, y))
        if y >= 1 and value == int_(board[x][y-1]) - 1:
            neighbors.append((x, y-1))
        for neighbor in neighbors:
            dfs(board, board_shape, neighbor, visited_coordinates)


def dfs_duplicates(board, board_shape, coordinates, visited_coordinates):
    x, y = coordinates
    if board[x][y] == '.':
        value = -1
    else:
        value = int_(board[x][y])
    neighbors = []
    visited_coordinates.append(coordinates)
    if x >= 1 and value == int_(board[x-1][y]) - 1:
        neighbors.append((x-1, y))
    if y <= board_shape[1] - 2 and value == int_(board[x][y+1]) - 1:
        neighbors.append((x, y+1))
    if x <= board_shape[0] - 2 and value == int_(board[x+1][y]) - 1:
        neighbors.append((x+1, y))
    if y >= 1 and value == int_(board[x][y-1]) - 1:
        neighbors.append((x, y-1))
    for neighbor in neighbors:
        dfs_duplicates(board, board_shape, neighbor, visited_coordinates)


def solve_part_1(board):
    score = 0
    board_shape = len(board), len(board[0])
    for i in range(board_shape[0]):
        for j in range(board_shape[1]):
            if board[i][j] == '0':
                visited_coordinates = set()
                dfs(board, board_shape, (i, j), visited_coordinates)
                for x, y in visited_coordinates:
                    if board[x][y] == '9':
                        score += 1
    return score


def solve_part_2(board):
    score = 0
    board_shape = len(board), len(board[0])
    for i in range(board_shape[0]):
        for j in range(board_shape[1]):
            if board[i][j] == '0':
                visited_coordinates = []
                dfs_duplicates(board, board_shape, (i, j), visited_coordinates)
                for x, y in visited_coordinates:
                    if board[x][y] == '9':
                        score += 1
    return score


with open('input.txt', 'r') as f:
    data = f.read().splitlines()

print(solve_part_1(data))
print(solve_part_2(data))
```
