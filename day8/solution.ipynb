```python
from collections import defaultdict


def find_all_antennas(city):
    antennas = defaultdict(list)
    for i in range(len(city)):
        for j in range(len(city[0])):
            if city[i][j] not in ['.', '#']:
                antennas[city[i][j]].append((i, j))
    return antennas


def find_antinode_pair(first_antenna, second_antenna, city_shape):
    first_x, first_y = first_antenna
    second_x, second_y = second_antenna
    x_diff = first_x - second_x
    y_diff = first_y - second_y
    firts_antinode = None
    second_antinode = None
    if 0 <= first_x + x_diff <= city_shape[0] - 1:
        if 0 <= first_y + y_diff <= city_shape[1] - 1:
            firts_antinode = first_x + x_diff, first_y + y_diff
    if 0 <= second_x - x_diff <= city_shape[0] - 1:
        if 0 <= second_y - y_diff <= city_shape[1] - 1:
            second_antinode = second_x - x_diff, second_y - y_diff
    return [firts_antinode, second_antinode]


def find_antinode_group(first_antenna, second_antenna, city_shape):
    first_x, first_y = first_antenna
    second_x, second_y = second_antenna
    x_diff = first_x - second_x
    y_diff = first_y - second_y
    antinode_x = first_x
    antinode_y = first_y
    antinode_group = [(first_x, first_y)]
    while antinode_x >= 0 and antinode_x <= city_shape[0] and antinode_y >= 0 and antinode_y <= city_shape[1]:
        antinode_x = antinode_x - x_diff
        antinode_y = antinode_y - y_diff
        if antinode_x >= 0 and antinode_x <= city_shape[0] - 1:
            if antinode_y >= 0 and antinode_y <= city_shape[1] - 1:
                antinode_group.append((antinode_x, antinode_y))
    antinode_x = first_x
    antinode_y = first_y
    while antinode_x >= 0 and antinode_x <= city_shape[0] and antinode_y >= 0 and antinode_y <= city_shape[1]:
        antinode_x = antinode_x + x_diff
        antinode_y = antinode_y + y_diff
        if antinode_x >= 0 and antinode_x <= city_shape[0] - 1:
            if antinode_y >= 0 and antinode_y <= city_shape[1] - 1:
                antinode_group.append((antinode_x, antinode_y))
    return antinode_group


def solve_part_1(antennas, city_shape):
    antinodes = set()
    for antenna_type in antennas.keys():
        for i in range(len(antennas[antenna_type])):
            for j in range(i+1, len(antennas[antenna_type])):
                first_antenna = antennas[antenna_type][i]
                second_antenna = antennas[antenna_type][j]
                current_antinode_pair = find_antinode_pair(first_antenna, second_antenna, city_shape)
                first_antinode, second_antinode = current_antinode_pair
                antinodes.add(first_antinode)
                antinodes.add(second_antinode)
    if None in antinodes:
        return len(antinodes) - 1
    return len(antinodes)


def solve_part_2(antennas, city_shape):
    antinodes = set()
    for antenna_type in antennas.keys():
        for i in range(len(antennas[antenna_type])):
            for j in range(i+1, len(antennas[antenna_type])):
                first_antenna = antennas[antenna_type][i]
                second_antenna = antennas[antenna_type][j]
                current_antinode_group = find_antinode_group(first_antenna, second_antenna, city_shape)
                for antinode in current_antinode_group:
                    antinodes.add(antinode)
    if None in antinodes:
        return len(antinodes) - 1
    return len(antinodes)


with open('input.txt', 'r') as f:
    data = f.read().splitlines()

antennas = find_all_antennas(data)
city_shape = len(data), len(data[0])
print(solve_part_1(antennas, city_shape))
print(solve_part_2(antennas, city_shape))
```
